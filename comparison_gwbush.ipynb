```python
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
from nltk.stem import PorterStemmer
```

```python
#nltk.download('punkt')

stop_words = set(stopwords.words('english')) 
porter = PorterStemmer()

tokens = []
for sent in all_text:
    word_tokens = word_tokenize(sent)   
    filtered_sentence = [porter.stem(w) for w in word_tokens if not w in stop_words] 
    tokens.append(filtered_sentence)
```

```python
from gensim import corpora, matutils
from gensim.models import TfidfModel
import numpy as np
```

```python
dictionary = corpora.Dictionary(tokens)

corpus = [dictionary.doc2bow(doc) for doc in tokens]
model = TfidfModel(corpus)
tfidf = [model[i] for i in corpus]
doc_matrix = matutils.corpus2csc(tfidf).transpose()
```

```python
c=len(tokens)
cos_sim = np.zeros([c, c]) # 初期化0の行列を作成
var_SDGs = doc_matrix.dot(doc_matrix.transpose()).toarray()

for j in range(c):
    cos_sim[-1,j] = var_SDGs[-1,j]/(np.sqrt(var_SDGs[-1, -1])*np.sqrt(var_SDGs[j, j]))

bush_sim =cos_sim[c-1:c,:-1] # input自身の要素だけ抜いた最後の行
```

```python
result = [i for i in list(bush_sim[0])]
```

```python
result = []
for i in list(bush_sim[0]):
    print('{0:.10f}'.format(i))
    result.append('{0:.10f}'.format(i))
```

```python
for i in range(len(result)):
    if float(result[i]) >0.5:
        print(result[i])
        print(i)
```
